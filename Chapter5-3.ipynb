# 데이터합치기

```python
import pandas as pd
```

## 열합치기

```python
날씨 = pd.read_excel('./data/weather.xlsx')
날씨[:3]
```

```python
수원날씨 = 날씨[날씨.location == 'suwon']
수원날씨[:3]
```

```python
식당 = pd.read_excel('./data/food_order.xlsx')
식당[:3]
```

식당과 수원날씨, 날짜가 공통적인 부분이 있네요. 날씨에 따른 주문기록의 연관성을 살펴보기 위해 합쳐보겠습니다.

```python
pd.merge(식당, 수원날씨)
```

data 열의 제목이 공통적이고, 그 값들도 공통적인 값이 있습니다. 공통적인 열 제목이 하나인 경우에 자동으로 해당열을 기준으로 합쳐집니다.

```python
pd.merge(식당, 수원날씨, on='date')[:3]
```

열 제목이 공통적이지 않으면???

```python
식당2 = 식당.rename(columns = {'date':'날짜'})
식당2[:3]
```

```python
# 서로 이름이 다른 열로 합치려면 열 이름을 지정해야 합니다.
pd.merge(식당2, 수원날씨, left_on="날짜", right_on="date")[:3]
```

```python
식당3 = 식당.set_index('date')
식당3[:3]
```

```python
pd.merge(식당3, 수원날씨, left_index=True, right_on='date')[:3]
```

## 축 이어 붙이기

가로또는 세로 방향으로 데이터 프레임을 이어붙입니다.

```python
NC13 = pd.read_excel('./data/NC Dinos 2013.xlsx')
NC13['시즌'] = 2013
NC14 = pd.read_excel('./data/NC Dinos 2014.xlsx')
NC14['시즌'] = 2014
NC15 = pd.read_excel('./data/NC Dinos 2015.xlsx')
NC15['시즌'] = 2015
NC13[:3]
```

색인 방향으로 "길게" 이어붙이기

```python
NC다이노스 = pd.concat([NC13, NC14, NC15])
```

```python
NC다이노스
```

```python
#시즌별, 선수별
NC다이노스.set_index(['시즌', '선수명'])
```
