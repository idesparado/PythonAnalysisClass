## 엑셀파일


파일은 크게 두 종류가 있습니다.
* 텍스트 : 어느 텍스트 편집기에서나 열 수 있고, 편집할 수 있습니다. (인코딩이 올바르다는 가정하에)
* 바이너리 : 해당 파일의 형식을 이해하는 전용 앱이 필요함. (엑셀, 워드 등)

```python
경로 = './data/koreaexim/international/Global+Market+Daily+Table%28값복사%29_0803.xls'
pd.read_excel(경로)
```

엑셀을 읽어오는 시점에서 최대한 영역을 잘 석택하겠습니다.
* C:K열을 선택합니다. parse_cols 인자를 통해 선택할 수 있습니다.
* 처음 3줄은 제외합니다. skiprows로 건너뛸 줄을 설정할 수 있습니다.
* 열제목은 각 표별로 설정해야 하기 때문에, 열제목 없음을 설정합니다.

```python
국제0803 = pd.read_excel(경로, parse_cols="C:K", sskiprow=3, header=None)
```

빈줄을 삭제하겠습니다.

데이터프레임.dropna() : NaN가 포함된 모든 색인을 삭제합니다.

NaN 기준을 설정할 필요가 있을 것 같습니다. 모든 값이 NaN인 경우에만 삭제하도록 하겠습니다.

```python
국제0803 = 국제0803.dropna(how='all')
```

```python
# 색인 번호가 고르지 않게 되어 재설정합니다.
국제0803 = 국제0803.reset_index(drop=True)
국제0803
```

```python
def 수출입은행국제(경로):
    # 엑셀파일 읽기
    원본 = pd.read_excel(경로, skiprow=3, parse_cols="C:K", header=None)
    
    # 빈줄 정리
    원본 = 원본.dropna(how='all')
    #빈 줄 정리로 고르지 않은 색인 재설정
    원본 = 원본.reset_index(drop=True)
    
    # 표 잘라내기
    표범위목록 = 표범위추출(원본)
    표목록 = []
    for 범위 in 표범위목록:
        시작, 끝 = 범위
        # 표의 길이가 2보다 작으면 제외
        if 끝 - 시작 < 2:
            continue
        
        표 = 원본[시작:끝]                
        표 = 표.reset_index(drop=True)
        # 색인제목 추출
        색인제목 = 표.ix[0,0]
        # 열제목 추출
        열제목 = 표.ix[1, 1:]
        열제목.name = '' # 기존의 값
        # 색인과 열제목 설정
        표 = 표[2:].set_index(0)
        표.index.name = 색인제목
        표.columns = 열제목
        
        표목록.append(표)
    return 표목록

def 표범위추출(프레임):
    # 표범위 = [(0,9), (9,17), (17,25), (25,30), (30,45)]
    제목인가 = 프레임.notnull().sum(1) == 1
    시작지점 = list(프레임[제목인가].index)
    끝지점 = 시작지점[1:] + [len(프레임)]
    표범위 = list(zip(시작지점, 끝지점))
    return 표범위
```
